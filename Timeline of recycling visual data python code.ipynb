```python
from matplotlib import pyplot as plt
```

```python
years = [1960, 1970, 1980, 1990, 2000, 2005, 2010, 2015, 2016, 2017]
```

```python
#Data obtained here: https://www.epa.gov/facts-and-figures-about-materials-waste-and-recycling/national-overview-facts-and-figures-materials
Paper_and_paperboard = [17, 15, 21, 28, 43, 50, 63, 67, 68, 66]
Glass = [2, 1, 5, 20, 23, 21, 27, 28, 28, 27]
Plastics = [0.5, 0.5, 1, 2, 6, 6, 8, 9, 9, 8]
Yard_trimmings = [0.5, 0.5, 0.5, 12, 2, 62, 58, 61, 66, 69]
Lead_acid_batteries = [0.5, 76, 70, 97, 93, 96, 99, 99, 99, 99]
```

```python
plt.figure(figsize=(15,9))
plt.plot (years, Paper_and_paperboard)
plt.plot (years, Glass)
plt.plot (years, Plastics)
plt.plot (years, Yard_trimmings)
plt.plot (years, Lead_acid_batteries)
plt.xlabel('Year')
plt.ylabel('Recycling as a percentage of waste generation in the United States')
plt.legend(['Paper and paperboard', 'Glass', 'Plastics', 'Yard trimmings', 'Lead acid batteries'])
plt.show()
```
